```python
def inpute_age(cols):
    Age = cols[0]
    Pclass = cols[1]

    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
df['Age'] = df[['Age', 'Pclass']].apply(inpute_age, axis=1)
```

```python
sns.heatmap(df.isnull(), yticklabels = False, cbar = False, cmap = 'viridis')
```

```python
df.drop('Cabin', axis=1, inplace=True)
```

```python
df.head(20)
```

```python
sex = pd.get_dummies(df['Sex'], drop_first=True)
embark = pd.get_dummies(df['Embarked'], drop_first=True)
```

```python
df.drop(['Sex','Embarked', 'Name', 'Ticket'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df = pd.concat([df, sex, embark], axis=1)
```

#Task 8: Perform Data-Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Survived',axis=1), df['Survived'], test_size=0.25, random_state=101) # X --> df.drop('Survived',axis=1), Y --> df['Survived']
```

#Task 9: Choose the ML Method to follow

```python
from sklearn.linear_model import LogisticRegression
```

#Task 10: Train the ML Model

```python
logmodel = LogisticRegression()
logmodel.fit(X_train, y_train)
```

#Task 11: Test the Model

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predictions))
```

```python
print(logmodel.score(X_test,y_test))
pp = logmodel.predict(X_test)
print("Survived", sum(pp!=0))
print("NOT Survived", sum(pp==0))
```
